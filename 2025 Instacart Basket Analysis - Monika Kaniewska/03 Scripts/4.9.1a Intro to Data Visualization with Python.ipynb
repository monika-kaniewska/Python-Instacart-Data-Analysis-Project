```python
df_dups = df_cust[df_cust.duplicated()]
```

```python
df_dups.shape
```

There are no duplicates in df_cust dataframe.

```python
# Finding mixed type data
for col in df_cust.columns.tolist():
    col_df = df_cust[[col]]
    col_types = col_df.apply(lambda x: x.map(type))
    first_type = df_cust[[col]].iloc[0].apply(type)
    weird = (col_types != first_type).any(axis=1)
    if len(df_cust[weird]) > 0:
        print(col)
```

The column first_name has mixed type of data.
