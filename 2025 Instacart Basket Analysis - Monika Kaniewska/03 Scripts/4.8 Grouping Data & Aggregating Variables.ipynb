```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price']>=10, 'spender type'] = 'High spender'
```

```python
# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.
```

The criteria for the flag are as follows:
median “days_since_prior_order” > 20, then “Non-frequent customer”;
median > 10 and <= 20, then “Regular customer”;
median <= 10, then “Frequent customer”

```python
# Creating new column with median of the “days_since_prior_order” column
df_ords_prods_merge['median_days_since_prior_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.head()
```

```python
# Creating customer flags based on median of the days_since _prior_order
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] > 20, 'order frequency'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_prior_order'] > 10) & (df_ords_prods_merge['median_days_since_prior_order'] <= 20), 'order frequency'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] <= 10, 'order frequency'] = 'Frequent customer'
```

```python
df_ords_prods_merge.head()
```
